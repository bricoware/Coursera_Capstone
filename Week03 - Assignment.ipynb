1. import the libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

2. Retrieve the content page from Wikipedia

```python
paxina = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
resultado = requests.get(paxina).text
```

3. Create the BeautifulSoup Object

```python
soup = BeautifulSoup(resultado, 'html.parser')
#print(soup.prettify())
```

4. Scrap the Wikipedia List of postal codes of Canada

```python
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
```

```python
# print(table_contents)
```

5. Create the Pandas Dataframe and replace the name of some Boroughs 

```python
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

1. Retrieving Toronto latitude and longitude

```python
address = 'Toronto, Ca'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

2. Drawing the map of Totonto with all the places

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(datos['Latitude'], datos['Longitude'], datos['Borough'], datos['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

3. Specifying Borough "Downtown Toronto" as the scope and saving all neighbourghs in a new DataFrame

```python
toronto_data = datos[datos['Borough'] == 'Downtown Toronto'].reset_index(drop=True)
toronto_data.head()
```
